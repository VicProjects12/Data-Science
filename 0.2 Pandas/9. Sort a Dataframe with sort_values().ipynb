```python
import pandas as pd
```

```python
# reading the csv files
df_exams = pd.read_csv('/Users/vic/Downloads/StudentsPerformance.csv')
df_exams
```

```python
# 1 Sort a dataframe
df_exams.sort_values(by='math score')
```

```python
# Sort by one column
```

```python
# sort by descending by one column
df_exams.sort_values(by='math score', ascending=False)
```

```python
df_exams
```

```python
# sort by descending by multiple columns
df_exams.sort_values(by=['math score', 'reading score'], ascending=False)
```

```python
# sort descending by multiple columns and update dataframe
df_exams.sort_values(by=['math score', 'reading score'], ascending=False,
                    inplace=True)
```

```python
df_exams
```

```python
# Sort descending with a key function
df_exams.sort_values('race/ethnicity', ascending=True,
                    key=lambda col:col.str.lower())
```
